### Task II

Create a batch of input vectors ùëã of shape (5, 3).\
Define learnable weights ùëä of shape (3,) and a bias ùëè (a scalar).\
Compute the output: ùë¶ = ùëãùëä^T + ùëè (what is its shape?).\
Define a target vector ùë¶ and a mean squared error loss: ùêø = mean((ùë¶ ‚àí ùë¶_target)^2).\
Perform backpropagation to compute gradients.

```python
X = torch.rand((5, 3))

W = torch.rand((3), requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)

y = X @ W.T + b

y.shape
```

```python
y_true = torch.rand((y.shape))
y_true, y
```

```python
def mean_squared_error(y_pred, y_true):
    return torch.mean((y_pred - y_true) ** 2)
```

```python
L = mean_squared_error(y, y_true)
L
```

backpropagation

```python
L.backward()
```

```python
W.grad, b.grad
```
